# Definir el Problema a Resolver

**Objetivo principal**

Mediante el uso de los registros de pacientes, que se encuentran en el data-set, determinar: *¿Qué pacientes tienen enfermedad hepática y cuáles no?*

**Descripción del problema**

Los pacientes con enfermedad hepática han aumentado continuamente debido al consumo excesivo de alcohol, la inhalación de gases nocivos, la ingesta de 
alimentos contaminados, los encurtidos y las drogas. 

Este conjunto de datos contiene 416 registros de pacientes hepáticos y 167 
registros de pacientes no hepáticos recopilados del noreste de Andhra Pradesh,
 India. La columna "Dataset" es una etiqueta de clase utilizada para dividir
 grupos en pacientes hepáticos (enfermedad hepática) o no (sin enfermedad). 
Este conjunto de datos contiene 441 registros de pacientes masculinos y 142 
registros de pacientes femeninos. Cualquier paciente cuya edad excedió 89 se enumera como la edad "90". 

Es un *`problema de clasificación`* (A pesar de que los registros en la variable objetivo son numéricos, en realidad son categorías siendo **UNO**: Tienen enfermedad hepática y **DOS**: No tienen enfermedad hepática) que se debe analizar por modelos de machine learning.
        
**Contextualización**

Para poder entender de qué se trata el contenido del dataset, empezamos analizando el funcionamiento del hígado, el cual permite: 
- Producción de bilis.
- Producción de proteinas, colesterol, y otros transportadores
- Producción de glucógeno desde exceso de glucosa, equilibrio y producción de glucosa
- Regulación niveles de aminoácidos (captación y procesamiento )
- Procesamiento de la hemoglobina para la distribución de su contenido de hierro.
- Conversión del amoníaco tóxico en urea
- Depuración de fármacos y otras sustancias tóxicas de la sangre
- Regulación de la coagulación sanguínea.
- Producción de factores de inmunidad y depuración de bilirrubina[1].  

Como vemos, el hígado es un órgano vital, al cumplir funciones de producción, regulación y filtración de sustancias orgánicas, hace relevante realizar este estudio, y más en el sentido de que han aumentado en las personas, los malos hábitos, por lo que el consumo de sustancias tóxicas, en exceso, puede sobrecargar el trabajo del hígado, reflejado en enfermedades como la hepatitis. 

Para poder determinar si una persona es propensa a una enfermedad hepática, los médicos ordenan, la realización de exámenes de sangre. Propiamente, estos exámenes llevan acabo análisis, llamados "Función Hepática". Los análisis de función hepática son análisis de sangre utilizados para diagnosticar y supervisar enfermedades o daño en el hígado. Miden los niveles de determinadas enzimas y proteínas en la sangre [2]. En este dataset, vemos en las columnas, los registros de dichos análisis, por lo cual hablaremos de ellos, junto con sus valores normales:


1. **Total_Bilirubin (TBIL) (Bilirubina total):** La bilirrubina es un derivado de la degradación de los glóbulos rojos. Por lo general, pasa a través del hígado y es eliminada del organismo. Pero si esto no ocurre, a causa de una enfermedad hepática, los niveles de bilirrubina en la sangre suelen aumentar y la piel se torna amarillenta (ictericia). El análisis de bilirrubina total, mide el nivel de toda la bilirrubina en la sangre. Permite detectar enfermedades como: Hepatopatías, incluida la hepatitis y obstrucción de la vesícula biliar; ictericia. Los niveles normales están entre 0,3 – 1,3 gramos por decilitro (g/dl) (Suma de billirubina directa y bilirubina indirecta).
2. **Direct_Bilirubin (DBIL) (Bilirubina directa)**: Mide solamente la bilirrubina que ha sido procesada por el hígado y como resultado se ha unido a otras sustancias químicas. Detecta enfermedades como: Hepatopatías, alteraciones hemolíticas, hematológicas y metabólicas, incluidas la hepatitis y la obstrucción de la vesícula biliar. Los niveles normales están entre 0,1 – 0,4 gramos por decilitro (g/dl).
3.**Alkaline_Phosphotase (ALP) (Fosfatasa Alcalina):** La ALP es una enzima que se encuentra en el hígado y los huesos y es importante para descomponer las proteínas. Detecta enfermedades como: Enfermedades del hígado, óseas, paratiroideas e intestinales. Los niveles normales están entre 40 a 129 Unidades/Litro (U/L)
4. **Alamine_Aminotransferase (ALT) (Alanina aminotransferasa):** La ALT es una enzima que se encuentra en el hígado y que ayuda a convertir las proteínas en energía para las células hepáticas. Su análisis permite la detección de enfermedades como: Hepáticas, incluidas la hepatitis viral y la cirrosis (Posible alto consumo de bebidas alcoholicas). Los niveles normales están entre 7 a 55 unidades por litro (U/L).
5. **Aspartate_Aminotransferase (AST) (Aspartato aminotransferasa):** Esta enzima, que desempeña un papel en el procesamiento de proteínas, está presente en el hígado, el corazón, los músculos y los riñones. Su análisis permite detectar: Hepatopatías, incluidas la hepatitis e ictericia viral, shock. Los niveles normales están entre 8 a 48 unidades por litro (U/L)
6. **Total_Protiens (TP) (Proteinas totales):** Se mide la cantidad total de todo tipo de proteínas en la sangre. Detecta enfermedades como: Hepatopatías, enfermedades del riñón o de la médula ósea, alteraciones metabólicas y de la nutrición. Los niveles normales están entre 6,3 a 7,9 gramos por decilitro (g/dl). (Suma de todas las proteinas).
7. **Albumin (ALB)(Albúmina):** La albúmina ayuda a mantener el líquido dentro del torrente sanguíneo sin que se filtre a otros tejidos. También transporta varias sustancias por el cuerpo, por ejemplo, hormonas, vitaminas y enzimas [5].Su estudio permite detectar: Patologías del hígado y del riñón. Los niveles normales están entre 3,5 a 5,0 gramos por decilitro (g/dl)
8. **Albumin_and_Globulin_Ratio (A/G) (Relación Albúmina - Globulina):** Es un análisis de sangre que mide los niveles de proteínas en su cuerpo. Su hígado produce la mayoría de las proteínas que están en su sangre.[6] Ambas son proteinas, por lo que realizar esta relación permite establecer en especial si tiene cirrosis y otras enfermedades hepáticas. En condiciones normales, esa relación es superior a 1,0 (No tiene unidad por ser una relación) [7]. Esto, debido a que por lo general la Globulina tiene menores valores (Rango normal entre 23 a 35 g/L [6] o sea 2.5 a 3.5 g/dl) versus la Albúmina.
9. **Age**: Es la edad de la persona que se le realizó los examenes. Tiene valores entre  0 y 90 años
10. **Gender**: Sexo de las persona a la que se le realizarom los examenes. Solo tiene dos categorías en la variable: Female (Mujer) y Male (Hombre).
11. **Dataset**: Identifica si la persona tiene o no enfermedad hepática. Solo tiene dos categorías en la variable: 1 (Tiene enfermedad hepática) y 2 (No tiene enfermedad hepática) (Debe convertirse a cero).

- **Significado de cada Variable** 
 1. **Total_Bilirubin (TBIL) (Bilirubina total):** Mide el nivel de toda la bilirrubina en la sangre. (Suma de billirubina directa y bilirubina indirecta).
 2. **Direct_Bilirubin (DBIL) (Bilirubina directa)**: Mide solamente la bilirrubina que ha sido procesada por el hígado y como resultado se ha unido a otras sustancias químicas.
 3.**Alkaline_Phosphotase (ALP) (Fosfatasa Alcalina):** Cantidad de la enzima ALP alojada en el hígado y huesos.
 4. **Alamine_Aminotransferase (ALT) (Alanina aminotransferasa):** Cantidad de la enzima ALT que hay en el hígado.
 5. **Aspartate_Aminotransferase (AST) (Aspartato aminotransferasa):** Cantidad de la enzima AST presente en el hígado, el corazón, los músculos y riñones.
 6. **Total_Protiens (TP) (Proteinas totales):** Mide la cantidad total de todo tipo de proteínas en la sangre.
 7. **Albumin (ALB)(Albúmina):** Mide la cantidad de la proteína albúmina dentro del torrente sanguíneo.
 8. **Albumin_and_Globulin_Ratio (A/G) (Relación Albúmina - Globulina):** Es una relación entre la proteína Albúmina y la proteína Globulina.
 9. **Age**: Es la edad de la persona que se le realizó los examenes
 10. **Gender**: Sexo de las persona a la que se le realizarom los examenes.
 11. **Dataset**: Identifica si la persona tiene o no enfermedad hepática.

```python
def summarize_classification(y_test, y_pred):
    
    acc = accuracy_score(y_test, y_pred, normalize=True)
    num_acc = accuracy_score(y_test, y_pred, normalize=False)

    prec = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    
    return {'accuracy': acc, 
            'precision': prec,
            'recall':recall, 
            'accuracy_count':num_acc}
```

```python
def build_model(classifier_fn,                
                name_of_y_col, 
                names_of_x_cols, 
                dataset, 
                test_frac=0.2):
    
    X = dataset[names_of_x_cols]
    Y = dataset[name_of_y_col]

    x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=test_frac)
       
    model = classifier_fn(x_train, y_train)
    
    y_pred = model.predict(x_test)

    y_pred_train = model.predict(x_train)
    
    train_summary = summarize_classification(y_train, y_pred_train)
    test_summary = summarize_classification(y_test, y_pred)
    
    pred_results = pd.DataFrame({'y_test': y_test,
                                 'y_pred': y_pred})
    
    model_crosstab = pd.crosstab(pred_results.y_pred, pred_results.y_test)
    
    return {'training': train_summary, 
            'test': test_summary,
            'confusion_matrix': model_crosstab}
```

```python
def compare_results():
    for key in result_dict:
        print('Classification: ', key)

        print()
        print('Training data')
        for score in result_dict[key]['training']:
            print(score, result_dict[key]['training'][score])

        print()
        print('Test data')
        for score in result_dict[key]['test']:
            print(score, result_dict[key]['test'][score])
       
        print()
```

**Nota:**

Por lo anterior, al ser de mayor impacto o relevancia, el falso negativo, se escoge como medida de evaluación de los modelos de ML, el **RECALL**. 

-  Como se explicó antes, se escogió el RECALL, por el hecho de que no tenemos balanceada la variable objetivo, así que no se usa el Accuracy, sino esta medida que trabaja con los Outliers.

```python
# Como mencionamos al principio de este módulo de Machine Learning, el número de  Cross-Validations son 10,
# pero hasta el momento no hemos empleado ese número de validaciones. Para aplicar la validación
# cruzada empleamos el siguiente código:

#Llamamos el modulo para poder seleccionar diferentes modulos a la hora de ejecutarlos
from sklearn import model_selection

seed =2
models = []

#Regresión logística
models.append(('LR', LogisticRegression(solver='liblinear')))

# SVM Linear
models.append(('SVM', SVC(C=1.0, kernel='linear', max_iter=1000, tol=1e-3)))

# SVM Kernel
models.append(('SVM', SVC(C=1.0, kernel='rbf', max_iter=1000, tol=1e-3)))

# Clasificador arboles de decisión
models.append(('CART', DecisionTreeClassifier()))

# Clasificador Random Forest
models.append(('CART', RandomForestClassifier()))

# K-NN KNeighborsClassifier
models.append(('neighbors', KNeighborsClassifier(n_neighbors=5)))

# Naïve Bayes
models.append(('NB', GaussianNB()))

# Evaluación de cada modelo con cross validation = 10 y medida de Recall
results = []
names = []
scoring = 'recall'
for name, model in models:

	# Kfol cross validation for model selection
	kfold = model_selection.KFold(n_splits=10, random_state=None)
 
	#X train , y train
	cv_results = model_selection.cross_val_score(model, x_train, y_train, cv=kfold, scoring=scoring)
	results.append(cv_results)
	names.append(name)
	msg = f"({name}, {cv_results.mean()}, {cv_results.std()}"
	print(msg)
```

```python
# Calculando el sensitivity o recall, comparando los resultados de la predicción versus los que estaban en la variable objetivo (test set 30%)
n, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
sensitivity = tp / (tp+fn)
sensitivity
```

```python
# Recordando que el mejor SVm fue lineal, sus hiperparámetros son:
# {'C': 0.1, 'penalty': 'l1'} con solver='liblinear'
#
lr = LogisticRegression(C=0.1, penalty='l1', solver='liblinear')
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
print(classification_report(y_test, y_pred))
```
